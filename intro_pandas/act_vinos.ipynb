```python
import pandas as pd
```

```python
winmag_url = 'https://gist.githubusercontent.com/clairehq/79acab35be50eaf1c383948ed3fd1129/raw/407a02139ae1e134992b90b4b2b8c329b3d73a6a/winemag-data-130k-v2.csv'

df_wine = pd.read_csv(winmag_url, engine='python', index_col=0)
display(df_wine)
```

```python
print("¿Cuantas variables tiene el dataframe?")
n_var = len(df_wine.dtypes)
print(f'  El dataframe tiene {n_var} variables (sin contar el índice)')

print("¿Qué variables tienen valores perdidos?")
for col in df_wine:
  print (f'  {col}') if df_wine[col].isna().any() else ""

var_types = df_wine.dtypes
print("¿Qué variables son numéricas?")
num_var = var_types[var_types != object].index
for var in num_var:
  print(f'  {var}')

print("¿Qué variables son cualitativas?")
qual_var = var_types[var_types == object].index
for var in qual_var:
  print(f'  {var}')
```

```python
# Auxiliar
print(df_wine['country'].unique())
```

```python
print("Dataframe con únicamente vinos europeos")
eu_countrys = ['Italy', 'Portugal', 'Spain', 'France', 'Germany', 'Austria', 'Hungary', 'Greece', 'Romania', 'Turkey', 'Czech Republic', 'Slovenia', 'Luxembourg', 'Croatia', 'Georgia', 'England', 'Serbia', 'Moldova', 'Bulgaria', 'Cyprus', 'Armenia', 'Switzerland', 'Bosnia and Herzegovina', 'Ukraine', 'Slovakia', 'Macedonia']
df_wine_eu = df_wine.loc[df_wine.country.isin(eu_countrys)]
display(df_wine_eu)
```

```python
print("Tabla con el menor, mayor y promedio del precio de la botella por país")
df_groups = df_wine.groupby('country')['price'].agg(['min', 'max', 'mean'])
df_groups.columns = ['min', 'max', 'mean']
display(df_groups)
df_groups.plot() 

print("¿De que país es la botella de menor precio?")
cheaper_wine = df_groups['min'].idxmin()
print(f'  {cheaper_wine}')
```

```python
print("¿Cuantos vinos hay con *aroma a fresa* entre otras consideraciones snobs que vienen en la descripción?")
snob_words = ["nuance", "undertone", "subtle", "elegant", "complex", "refined", "bouquet", "ethereal", "lingering", "velvety", "opulent", "polished", "sublime"]
snobs_counter = df_wine['description'].str.count("(" + "|".join(snob_words) + ")").sum()
print(f'    Hay {snobs_counter} snobs')
```

```python
print(f'¿Cuantas designaciones diferentes hay?')
num_designation = df_wine['designation'].nunique()
print(f'  {num_designation}')

frec_designation = df_wine['designation'].value_counts()

print(f'¿Cuál es la más repetida?')
label_frec_designation = df_wine['designation'].value_counts().idxmax()
print(f'  {label_frec_designation}')

print(f'¿Cuantas veces se repite?')
num_frec_designation = df_wine['designation'].value_counts().max()
print(f'  {num_frec_designation}')
```

```python
print("Dataframe con las variables de variedad, país y precio (menores a 20$ dolares)")
df_wine_subset = df_wine.loc[df_wine.price < 20, ['variety', 'country', 'price']]
display(df_wine_subset)
```

```python
print("Tabla con la cantidad de vinos diferentes de la variedad *Pinot Noir* por cada país")
pinotnoir_freq_country = df_wine.groupby('country')['variety'].apply(lambda x: (x == 'Pinot Noir').sum())
print(pinotnoir_freq_country)
```
